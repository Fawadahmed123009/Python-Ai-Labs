```python
# Exercise Q1
strlist=list(input("Enter a word to reverse "))
```

```python
print("Before reverse : ",strlist)
```

```python
size=len(strlist)
```

```python
x=int(size/2)
j=size-1
for i in range(x):
  y=strlist[i]
  z=strlist[j]
  strlist[j]=y
  strlist[i]=z
  j=j-1

```

```python
print("After reverse : ",strlist)
```

```python
#Exercise Q2
s = "A man, a plan, a canal: Panama"
```

```python
def isPalindrome(s):
  return s==s[::-1]
```

```python
s=s.replace(",","")
s=s.replace(".","")
s=s.replace(" ","")
s=s.replace(",","")
s=s.replace(":","")
s=s.lower()
print(s)
```

```python
check=isPalindrome(s)
if check:
  print("It is a palindrome")
else:
  print("It is not a palindrome")
```

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd

data=pd.read_csv('iris.csv',header=None).values

x = []
y = []
with open('iris.csv','r') as csvfile:
  plots = csv.reader(csvfile, delimiter =',')

  for row in plots:
    x.append(row[0])
    y.append(row[1])


plt.bar(x, y, color = 'g', width = 0.5, label= "length")
plt.xlabel('length')
plt.ylabel('width')
plt.title('Sizes')
plt.legend()
plt.show()
```
